```python
import pandas as pd
```

```python
vine_df = pd.read_csv("vine_table.csv")
vine_df
```

```python
filtered_df = vine_df[vine_df.total_votes>20]
filtered_df
```

```python
helpful_df = filtered_df[(filtered_df.helpful_votes/filtered_df.total_votes>0.5)]
helpful_df
```

```python
paid_df = helpful_df[helpful_df.vine=="Y"]
paid_df
```

```python
unpaid_df = helpful_df[helpful_df.vine=="N"]
unpaid_df
```

```python
# Paid Reviews
Paid5Count = sum(paid_df.star_rating == 5)
PaidTotal = paid_df["star_rating"].count()
PaidPercent = Paid5Count/PaidTotal
```

```python
print("Paid 5 Star Reviews = ",Paid5Count)
print("Total Paid Reviews = ", PaidTotal)
print("Paid 5 Star Review % = ", PaidPercent)
```

```python
# Unpaid Reviews
unPaid5Count = sum(unpaid_df.star_rating == 5)
unPaidTotal = unpaid_df["star_rating"].count()
unPaidPercent = unPaid5Count/unPaidTotal
```

```python
print("Unaid 5 Star Reviews = ",unPaid5Count)
print("Total Unpaid Reviews = ", unPaidTotal)
print("Unpaid 5 Star Review % = ", unPaidPercent)
```
